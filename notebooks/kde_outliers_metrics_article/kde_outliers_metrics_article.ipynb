```python
import numpy as np
import matplotlib.pyplot as plt

from kdelearn.kde import KDE
from kdelearn.kde_funcs import KDEOutliersDetector
from kdelearn.bandwidth_selection import normal_reference, direct_plugin
```

```python
def func(x_train):
    bandwidth = direct_plugin(x_train, stage=2)
    kde = KDE().fit(x_train, bandwidth=bandwidth)
    scores = kde.pdf(x_train)

    metrics_list = []
    r_list = []
    for r in np.arange(0.01, 0.51, 0.005):
        r = round(r, 3)
        threshold = np.quantile(scores, r)
        labels_pred = np.where(scores <= threshold, 1, 0)

        n_outliers = labels_pred.sum()
        if n_outliers == 0:
            continue

        scores_o = scores[labels_pred==1]
        scores_i = scores[labels_pred==0]
        metrics = np.sum(scores_o) / np.sum(np.sort(scores_i)[:n_outliers])

        metrics_list.append(metrics)
        r_list.append(r)

    plt.plot(r_list, metrics_list, "-o")
    plt.xlabel("r")
    plt.ylabel("PI")
    plt.grid()
    plt.show()

    # Find minima
    iters = len(metrics_list)
    min_metrics_list = []
    min_r_list = []
    for i in range(1, iters - 1):
        m_prev, m_curr, m_next = metrics_list[i-1], metrics_list[i], metrics_list[i+1]
        if m_prev > m_curr < m_next:
            min_metrics_list.append(metrics_list[i])
            min_r_list.append(r_list[i])
    
    first_min = (min_metrics_list[0], min_r_list[0])
    idx = np.argmin(min_metrics_list)
    global_min = (min_metrics_list[idx], min_r_list[idx])
    
    print(f"r - pierwsze minimum: {first_min[1]:.3f}")
    print(f"r - globalne minimum: {global_min[1]:.3f}")
    
    return first_min, global_min, scores
```

```python
all_first_mins = []
all_global_mins = []

for a in np.arange(0.0, 0.51, 0.01):
    print(f"a = {round(a, 2)}")
    np.random.seed(1)

    # Data
    m_train = 1000
    mean1 = [0, 0]
    cov1 = [[1, 0], [0, 1]]
    m_train1 = int((1 - a) * m_train)
    x_train1 = np.random.multivariate_normal(mean1, cov1, size=m_train1)
    mean2 = [0, 0]
    cov2 = [[100, 0], [0, 100]]
    m_train2 = int(a * m_train)
    x_train2 = np.random.multivariate_normal(mean2, cov2, size=m_train2)
    x_train = np.concatenate((x_train1, x_train2))
    
    # Run
    first_min, global_min, scores = func(x_train)
    
    plt.figure(figsize=(6, 3.8))
    threshold = np.quantile(scores, first_min[1])
    labels_pred = np.where(scores <= threshold, 1, 0)
    plt.scatter(x_train[:,0], x_train[:,1])
    plt.scatter(x_train[labels_pred==1,0], x_train[labels_pred==1,1], c="r")
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.show()
    
    all_first_mins.append(first_min)
    all_global_mins.append(global_min)
```

```python
all_first_r = [r[1] for r in all_first_mins]
all_global_r = [r[1] for r in all_global_mins]

plt.scatter(all_first_r, all_global_r, zorder=2)
plt.xlim(0.0, 0.51)
plt.ylim(0.0, 0.51)
plt.xlabel("r (pierwsze minimum)")
plt.ylabel("r (globalne minimum)")
plt.grid()
plt.show()
```
