```python
import numpy as np
import matplotlib.pyplot as plt
from kdelearn import uniform, gaussian, epanechnikov, cauchy
```

```python
x_plot = np.linspace(-3, 3, 500)

plt.figure(figsize=(8, 5))
plt.plot(x_plot, uniform(x_plot), label='uniform')
plt.plot(x_plot, gaussian(x_plot), label='gaussian')
plt.plot(x_plot, epanechnikov(x_plot), label='epanechnikov')
plt.plot(x_plot, cauchy(x_plot), label='cauchy')
plt.legend()
plt.grid()
plt.show()
```
