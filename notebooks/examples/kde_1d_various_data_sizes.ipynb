```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from kdelearn.kde import KDE
```

```python
def foo(kernel_name, seed):
    np.random.seed(seed)

    fig, axs = plt.subplots(2, 3, figsize=(14, 8))

    for e, m in enumerate([10 ** i for i in range(1, 7)]):
        x_train = np.random.normal(0, 1, size=(m, 1))
        x_test = np.linspace(-4, 4, 100)[:, None]

        kde = KDE(kernel_name).fit(x_train)
        scores = kde.pdf(x_test)
        true_dist = norm.pdf(x_test)

        axs[e//3, e%3].set_title(f"m = {m:.0e}")
        axs[e//3, e%3].plot(x_test, true_dist, label="true dist.")
        axs[e//3, e%3].plot(x_test, scores, label="kde")
        axs[e//3, e%3].legend()
        axs[e//3, e%3].set_xlim(-4, 4)
        axs[e//3, e%3].set_ylim(0, 0.5)
        axs[e//3, e%3].grid()

    plt.show()
```

## uniform kernel

```python
foo(kernel_name="uniform", seed=0)
```

## gaussian kernel

```python
foo(kernel_name="gaussian", seed=0)
```

## epanechnikov kernel

```python
foo(kernel_name="epanechnikov", seed=0)
```

## cauchy kernel

```python
foo(kernel_name="cauchy", seed=0)
```
