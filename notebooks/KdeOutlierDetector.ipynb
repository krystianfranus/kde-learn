```python
import numpy as np
import matplotlib.pyplot as plt
from kdelearn.kde_funcs import KDEOutliersDetector
```

```python
np.random.seed(0)
x_train = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], size=1000)
# weights_train = np.random.normal(0, 1, size=(1000,)) + 10
# weights_train = None
x_test = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], size=1000)
```

```python
outlier_detector = KDEOutliersDetector("gaussian").fit(x_train)
outliers = outlier_detector.predict(x_test)
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1])
plt.scatter(x_test[outliers==1, 0], x_test[outliers==1, 1], c="red")
plt.show()
```
