```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from kdelearn import Kde
```

```python
np.random.seed(0)
```

```python
def foo(m):
    x_train = np.random.normal(size=(m, 1))
    kde = Kde().fit(x_train)

    x_plot = np.linspace(-4, 4, 100)[:,None]
    scores = kde.score_samples(x_plot)

    plt.figure(figsize=(8, 5))
    plt.plot(x_plot, norm.pdf(x_plot), label='standard normal dist')
    plt.plot(x_plot, scores, label='estimation')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
foo(m=10)
```

```python
foo(m=1_000)
```

```python
foo(m=1_000_000)
```
