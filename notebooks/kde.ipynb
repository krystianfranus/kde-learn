```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from kdelearn import Kde
```

```python
np.random.seed(0)
```

# Estimated bandwidth

```python
def foo(m, bandwidth=None):
    x_train = np.random.normal(size=(m, 1))
    kde = Kde().fit(x_train, bandwidth=bandwidth)

    x_plot = np.linspace(-4, 4, 100)[:,None]
    scores = kde.score_samples(x_plot)

    plt.figure(figsize=(8, 5))
    plt.plot(x_plot, norm.pdf(x_plot), label='standard normal')
    plt.plot(x_plot, scores, label='estimation')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
foo(m=10)
```

```python
foo(m=1000)
```

```python
foo(m=1_000_000)
```

```python
# foo(m=10_000_000)  # high memory usage!
```

# Fixed bandwidth

```python
bandwidth = np.array([0.5])
```

```python
foo(m=10, bandwidth=bandwidth)
```

```python
foo(m=1000, bandwidth=bandwidth)
```

```python
foo(m=1_000_000, bandwidth=bandwidth)
```
