```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from kdelearn.kde import KDE
```

```python
def foo(m):
    np.random.seed(0)

    x_train = np.random.normal(size=(m, 1))
    x_test = np.linspace(-4, 4, 100)[:, None]
    
    kde = KDE().fit(x_train)
    scores = kde.pdf(x_test)

    plt.figure(figsize=(8, 5))
    plt.plot(x_test, norm.pdf(x_test), label="standard normal")
    plt.plot(x_test, scores, label="kde")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
foo(m=100)
```

```python
foo(m=100_000)
```
