```python
data = pd.read_csv("../../data/pima/diabetes.csv")
```

```python
data.head()
```

```python
data["Outcome"].value_counts()
```

# Pregnancies

```python
plt.figure(figsize=(6 ,4))
plt.scatter(data.loc[data["Outcome"] == 0, "Pregnancies"], data.loc[data["Outcome"] == 0, "Pregnancies"]*0, alpha=0.2)
plt.scatter(data.loc[data["Outcome"] == 1, "Pregnancies"], data.loc[data["Outcome"] == 1, "Pregnancies"]*0, c="red", alpha=0.2)
plt.xlabel("Pregnancies")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data.loc[data["Outcome"] == 0, "Pregnancies"], bins=50, alpha=0.5)
plt.hist(data.loc[data["Outcome"] == 1, "Pregnancies"], bins=50, alpha=0.5, color="red")
plt.xlabel("Pregnancies")
plt.show()
```
