# 1d

## Orbital Period [days]

Historam

```python
plt.hist(data["pl_orbper"], bins=50)
plt.xlabel("orbinal period")
plt.show()
```

\newpage

Histogram z logarystmiczną skalą na osi y

```python
plt.hist(data["pl_orbper"], bins=50)
plt.yscale("log")
plt.xlabel("orbinal period")
plt.show()
```

\newpage

Histogram zlogarytmowanych wartości "orbital period" + logaryticzna skala na osi y

```python
plt.hist(np.log(data["pl_orbper"]), bins=50)
plt.yscale("log")
plt.xlabel("log (orbinal period)")
plt.show()
```

\newpage

Zwykły histogram, ale ograniczonej wartości od 0 do 200

```python
plt.hist(data["pl_orbper"], bins=50, range=(0, 200))
plt.xlabel("orbinal period")
plt.show()
```

```python
data["pl_orbper"].isna().sum()
```

```python
data["pl_orbper"].median()
```

```python
data[~data["pl_orbper"].isna()].sort_values("pl_orbper")[["pl_name", "pl_orbper"]][:10]
```

```python
data[~data["pl_orbper"].isna()].sort_values("pl_orbper")[["pl_name", "pl_orbper"]][-10:]
```
